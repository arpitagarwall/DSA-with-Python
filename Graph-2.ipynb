```python
#Kruskal's Algorithm

#defining edge class
class Edge:
    #creating init function to initialise constructor
    def __init__(self,src,dest,wt):
        self.src = src
        self.dest = dest
        self.wt = wt
        
def getParent(v,parent):
    #checking if vertex parent is vertext itself or not
    if(v == parent[v]):
        return v
    #else using induction hypothesis recursiion will give top most parent of vertex
    return getParent(parent[v], parent)

def kruskal(edges, nVertices):
    #creating parent array
    parent = [i for i in range(nVertices)]
    #sorting edge on basis o weight
    edges = sorted(edges, key = lambda edge:edge.wt)
    #initialising variables
    count = 0
    output = []
    i = 0
    #using while till count is less than nVertices - 1
    while count < (nVertices - 1):
        #assigning currentEdge with edges array at index i
        currentEdge = edges[i]
        #finding source and destination parents
        srcParent = getParent(currentEdge.src, parent)
        destParent = getParent(currentEdge.dest, parent)
        #checking if source and destination parent are not equal
        if srcParent != destParent:
            #appending that edge in ouput
            output.append(currentEdge)
            #increasing count by 1
            count = count + 1
            #changing source parent with destination
            parent[srcParent] = destParent
        #increasin i by 1
        i = i + 1
    return output

li = [int(ele) for ele in input().split()]
n = li[0]
E = li[1]
edges = []

for i in range(E):
    curr_input = [int(ele) for ele in input().split()]
    src = curr_input[0]
    dest = curr_input[1]
    wt = curr_input[2]
    edge = Edge(src,dest,wt)
    edges.append(edge)
    
output = kruskal(edges,n)
for edge in output:
    if edge.src < edge.dest:
        print(str(edge.src) + " " + str(edge.dest) + " " + str(edge.wt))
    else:
        print(str(edge.dest) + " " + str(edge.src) + " " + str(edge.wt))
```

```python
#Prim's Algorithm Problem

import queue
import sys
from sys import setrecursionlimit
setrecursionlimit(10**6)

#Defining new class as Graph
class Graph:
    #init method to initialise the constructor and making adjMatrix for nVertices
    def __init__(self, nVertices):
        self.nVertices = nVertices
        self.adjMatrix = [[0 for i in range(nVertices)] for j in range(nVertices)]
        
    def addEdge(self, v1, v2, wt):
        #changing 0 to wt at index v1 and v2 to show edge is added
        self.adjMatrix[v1][v2] = wt
        self.adjMatrix[v2][v1] = wt
        
    def __bfsHelper(self,sv,visited):
        #creating a queue q
        q = queue.Queue()
        #putiing element in the queue
        q.put(sv)
        #marking that element in visited as true so that it will not be visited again
        visited[sv] = True
        #till while loop is not empty
        while not q.empty():
            #getting the front element of queue assigning it to front node
            currentNode = q.get()
            #printing the current node
            print(currentNode, end=" ")
            #using for loop in range of nVertices
            for i in range(self.nVertices):
                #checking if adjMatrix at index[currentNode][i] is greater then 0 and visited[i] is false
                if self.adjMatrix[currentNode][i] > 0 and visited[i] is False:
                    #putting i in q
                    q.put(i)
                    #marking visited[i] as True
                    visited[i] = True
    def bfs(self):
        #creating visited array with all false value
        visited = [False for i in range(self.nVertices)]
        #running for loop for range nVertices
        for i in range(self.nVertices):
            #checking at visited[i] is false or not
            if visited[i] is False:
                #calling __bfsHelper
                self.__bfsHelper(i,visited)
        
    def removeEdge(self, v1, v2):
        #checking if there are edges in adjMatrix or not
        if self.containEdge(v1,v2) is False:
            return
        #assigning at that index equal to 0
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0
        
    def containEdge(self, v1, v2):
        #returning true if adjMatrix at index v1 and v2 is greater than 0
        return True if self.adjMatrix[v1][v2] > 0 else False
    
    def __str__(self):
        #for printing adjMatrix
        return str(self.adjMatrix)
    
    def __getMinVertex(self,visited, weight):
        #assigning min_vertex = -1
        min_vertex = -1
        for i in range(self.nVertices):
            if(visited[i] is False and (min_vertex == -1 or weight[min_vertex] > weight[i])):
                min_vertex = i
                
        return min_vertex
    
    def prims(self):
        #maintaining 3 arrays visited, parent and weight
        visited = [False for i in range(self.nVertices)]
        parent = [-1 for i in range(self.nVertices)]
        weight = [sys.maxsize for i in range(self.nVertices)]
        #assigning 0 weight for vertex 0
        weight[0] = 0
        #looping till nVertices - 1
        for i in range(self.nVertices - 1):
            #getting minimum vertex which is not visited
            min_vertex = self.__getMinVertex(visited, weight)
            #marking min_vertex as true in visited array
            visited[min_vertex] = True
            #looping till nVertices
            for j in range(self.nVertices):
                #if weight of vertex at index min_vertex and j is greater than 0 and visited at index j is false
                if self.adjMatrix[min_vertex][j] > 0 and visited[j] is False:
                    #if weight[j] is greater than adjMatrix[min_vertex][j] then assigning adjMatrix[min_vertex][j] to weight[j]
                    if(weight[j] > self.adjMatrix[min_vertex][j]):
                        weight[j] = self.adjMatrix[min_vertex][j]
                        #assigning min_vertex to parent[j]
                        parent[j] = min_vertex
        #printing the edges starting from 1                
        for k in range(1, self.nVertices):
            if k < parent[k]:
                print(str(k) + " " + str(parent[k]) + " " + str(weight[k]))
            else:
                print(str(parent[k]) + " " + str(k) + " " + str(weight[k]))
# Main
li = [int(ele) for ele in input().split()]
V = int(li[0])
E = int(li[1])

g = Graph(V)

for i in range(E) :
    arr = [int(ele) for ele in input().split()]
    g.addEdge(arr[0],arr[1],arr[2])

g.prims()    
```

```python
#Dijkstra's Algorithm

import queue
import sys
from sys import setrecursionlimit
setrecursionlimit(10**6)

#Defining new class as Graph
class Graph:
    #init method to initialise the constructor and making adjMatrix for nVertices
    def __init__(self, nVertices):
        self.nVertices = nVertices
        self.adjMatrix = [[0 for i in range(nVertices)] for j in range(nVertices)]
        
    def addEdge(self, v1, v2, dist):
        #changing 0 to wt at index v1 and v2 to show edge is added
        self.adjMatrix[v1][v2] = dist
        self.adjMatrix[v2][v1] = dist
        
    def __bfsHelper(self,sv,visited):
        #creating a queue q
        q = queue.Queue()
        #putiing element in the queue
        q.put(sv)
        #marking that element in visited as true so that it will not be visited again
        visited[sv] = True
        #till while loop is not empty
        while not q.empty():
            #getting the front element of queue assigning it to front node
            currentNode = q.get()
            #printing the current node
            print(currentNode, end=" ")
            #using for loop in range of nVertices
            for i in range(self.nVertices):
                #checking if adjMatrix at index[currentNode][i] is greater then 0 and visited[i] is false
                if self.adjMatrix[currentNode][i] > 0 and visited[i] is False:
                    #putting i in q
                    q.put(i)
                    #marking visited[i] as True
                    visited[i] = True
    def bfs(self):
        #creating visited array with all false value
        visited = [False for i in range(self.nVertices)]
        #running for loop for range nVertices
        for i in range(self.nVertices):
            #checking at visited[i] is false or not
            if visited[i] is False:
                #calling __bfsHelper
                self.__bfsHelper(i,visited)
        
    def removeEdge(self, v1, v2):
        #checking if there are edges in adjMatrix or not
        if self.containEdge(v1,v2) is False:
            return
        #assigning at that index equal to 0
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0
        
    def containEdge(self, v1, v2):
        #returning true if adjMatrix at index v1 and v2 is greater than 0
        return True if self.adjMatrix[v1][v2] > 0 else False
    
    def __str__(self):
        #for printing adjMatrix
        return str(self.adjMatrix)
    
    def __getMinVertex(self,visited, dist):
        #assigning min_vertex = -1
        min_vertex = -1
        for i in range(self.nVertices):
            if(visited[i] is False and (min_vertex == -1 or dist[min_vertex] > dist[i])):
                min_vertex = i
                
        return min_vertex
    
    def dijsktra(self):
        #maintaining 2 arrays visited and dist
        visited = [False for i in range(self.nVertices)]
        dist = [sys.maxsize for i in range(self.nVertices)]
        #assigning 0 distance for vertex 0
        dist[0] = 0
        #looping till nVertices - 1
        for i in range(self.nVertices - 1):
            #getting minimum vertex which is not visited
            min_vertex = self.__getMinVertex(visited, dist)
            #marking min_vertex as true in visited array
            visited[min_vertex] = True
            #looping till nVertices
            for j in range(self.nVertices):
                #if distance of vertex at index min_vertex and j is greater than 0 and visited at index j is false
                if self.adjMatrix[min_vertex][j] > 0 and visited[j] is False:
                    #if dist[j] is greater than self.adjMatrix[min_vertex][j] + dist[min_vertex] 
                    #then we neet to update dist[j] 
                    if dist[j] > self.adjMatrix[min_vertex][j] + dist[min_vertex]:
                        dist[j] = self.adjMatrix[min_vertex][j] + dist[min_vertex]
        #printing minimum distance                
        for k in range(self.nVertices):
            print(str(k) + " " + str(dist[k]))
# Main
li = [int(ele) for ele in input().split()]
V = int(li[0])
E = int(li[1])

g = Graph(V)

for i in range(E) :
    arr = [int(ele) for ele in input().split()]
    g.addEdge(arr[0],arr[1],arr[2])

g.dijsktra()    
```
