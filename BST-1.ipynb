```python
#Search In BST
import queue

class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def searchInBST(root, k):
    #base case if root is none then return
    if root == None:
        return
    #edge case if root.data == k then return true
    if root.data == k:
        return True
    #if k is greater then root then we will serach right subtree
    if root.data < k:
        #Assuming induction hypothesis recursion will give true if k is present in right subtree and we will return that
        return searchInBST(root.right, k)
    #if k is smaller then root then we will serach left subtree
    if root.data > k:
        #Assuming induction hypothesis recursion will give true if k is present in left subtree and we will return that
        return searchInBST(root.left, k)


def buildLevelTree(levelorder):
    index = 0
    length = len(levelorder)
    if length<=0 or levelorder[0]==-1:
        return None
    root = BinaryTreeNode(levelorder[index])
    index += 1
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currentNode = q.get()
        leftChild = levelorder[index]
        index += 1
        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)
        rightChild = levelorder[index]
        index += 1
        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)
    return root

# Main
levelOrder = [int(i) for i in input().strip().split()]
root = buildLevelTree(levelOrder)
k=int(input())
ans = searchInBST(root, k)
if ans:
    print("true")
else:
    print("false")
```

```python
#Construct BST

import queue
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def constructBST(lst):
    #maintaing length of given list
    list_length = len(lst)
    #base case if length of list is 0 then return
    if list_length == 0:
        return
    #mainting index of middle element in list
    mid_element = (list_length - 1) // 2
    #creating root node from middle element in list
    root = BinaryTreeNode(lst[mid_element])
    #maintianing left and right of middle elemnt from list
    left_lst = lst[0:mid_element]
    right_lst = lst[mid_element+1:]
    
    #assuming induction hypothesis recurrsion will give left and right subtree
    leftNode = constructBST(left_lst)
    rightNode = constructBST(right_lst)
    #joining root.left with left subtree and root.right with right subtree
    root.left = leftNode
    root.right = rightNode
    #returning root
    return root

def preOrder(root):
    # Given a binary tree, print the preorder traversal of given tree. Pre-order
    # traversal is: Root LeftChild RightChild
    if root==None:
        return
    print(root.data, end=' ')
    preOrder(root.left)
    preOrder(root.right)

# Main
n=int(input())
if(n>0):
    lst=[int(i) for i in input().strip().split()]
else:
    lst=[]
root=constructBST(lst)
preOrder(root)
```
