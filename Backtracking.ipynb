```python
#N-Queens

def isSafe(row, col, n, solution):
    #to check if 1 is present in vertically upwards we only decrease row number
    i = row - 1
    #while i is greater than eqaul to 0
    while i >= 0:
        #if in solution at index[i][col] 1 is present then return false
        if solution[i][col] == 1:
            return False
        i = i - 1
    #to check if 1 is present in diagonally left we decrease row and col number
    i = row - 1
    j = col - 1
    #while i and j is greater than eqaul to 0
    while i >= 0 and j >= 0:
        #if in solution at index[i][j] 1 is present then return false
        if solution[i][j] == 1:
            return False
        i = i -1
        j = j - 1
    #to check if 1 is present in diagonally right we only decrease row and increase col number    
    i = row - 1
    j = col + 1
    #while i is greater than eqaul to 0 and j is less than n
    while i >= 0 and j < n:
        #if in solution at index[i][j] 1 is present then return false
        if solution[i][j] == 1:
            return False
        i = i -1
        j = j + 1
    
    return True

def nQueenHelper(row,n,solution):
    #if row is equal to n then printing the solution
    if row == n:
        for i in range(n):
            for j in range(n):
                print(solution[i][j], end = " ")
        print()
        return
    #for every column in solution
    for col in range(n):
        #checking if isSafe method return true
        if isSafe(row, col, n, solution) is True:
            #assigning 1 at index [row][col]
            solution[row][col] = 1
            #assuiming induction hypothesis we will try to place 1 at each row in solution
            nQueenHelper(row+1,n,solution)
            #replacing 1 with 0 becuase we need to back track correctly
            solution[row][col] = 0
            
    return

def nQueen(n):
    
    solution = [[0 for j in range(n)] for i in range(n)]
    nQueenHelper(0,n,solution)
    
n = int(input())
nQueen(n)
```

```python
#Sudoku Solver
def isSafe(row,col,num,n,board):
    #checking in current column if there is num exist or not
    for i in range(n):
        if board[i][col] == num:
            return False
    #checking in current row if there is num exist or not
    for j in range(n):
        if board[row][j] == num:
            return False
    
    startRow = row - row % 3
    startCol = col - col % 3
    #checking in current box of 9 values if there is num exist or not
    for i in range(3):
        for j in range(3):
            if board[i + startRow][j + startCol] == num:
                return False
    return True

def solveSudokuHelper(row,col,n,board):
    #checking row and col it its equal to last column and last row then returning true
    if row == n - 1 and col == n - 1:
        return True
    #if col is eqaul to last column
    if col == n:
        #then making col = 0 for next row
        col = 0
        #moving to next row
        row = row + 1
    #if number at index [row][col] in board is greater than 0
    if board[row][col] > 0:
        #calling sudokuhelper again by increasing column by 1
        return solveSudokuHelper(row,col+1,n,board)
    #using for loop in range 1 to n+1
    for num in range(1,n+1,1):
        #calling other function isSafe and checking it is true or not
        if isSafe(row,col,num,n,board) is True:
            #then adding num at the index [row][col]
            board[row][col] = num
            #then checking if solveSudokuHelper() return true
            if solveSudokuHelper(row,col,n,board) is True:
                #then return true
                return True
        #for backtracking replacing 0 to previous element
        board[row][col] = 0
    
    return False
        

def solveSudoku(board):
    #maintaining length of board in variable n
    n = len(board)
    #calling helper function solveSudokuHelper()
    ans = solveSudokuHelper(0,0,n,board)
    #returning ans
    return ans

board = [[int(ele) for ele in input().split() ] for i in range(9)]
ans = solveSudoku(board)
if ans is True:
    print('true')
else:
    print('false')
```
