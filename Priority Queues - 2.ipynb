```python
#Inplace Heap Sort
def down_heapify(arr, i, n):
    #maintaining parentIndex, leftChildIndex, rightChildIndex
    parentIndex = i
    leftChildIndex = 2 * parentIndex + 1
    rightChildIndex = 2 * parentIndex + 2
    #while loop will run till leftChildIndex is less than n
    while leftChildIndex < n:
        #assigning parentIndex to minIndex
        minIndex = parentIndex
        #if element at minIndex is greater than element at leftChildIndex
        if arr[minIndex] > arr[leftChildIndex]:
            minIndex = leftChildIndex
        #if rightChildIndex is less than n element at minIndex is greater than element at rightChildIndex
        if rightChildIndex < n and arr[minIndex] > arr[rightChildIndex]:
            minIndex = rightChildIndex
        #when minIndex == parentIndex we break the loop 
        if minIndex == parentIndex:
            return
        #swapping element 
        arr[minIndex], arr[parentIndex] = arr[parentIndex], arr[minIndex]
        #assigning minIndex to parentIndex
        parentIndex = minIndex
        #calculating leftChildIndex and rightChildIndex
        leftChildIndex = 2*parentIndex+1
        rightChildIndex = 2*parentIndex+2
    return


def heapSort(arr):
    #calling down_heapify till n // 2 because we have to work only on non-leaf nodes
    for i in range(n//2 -1, -1, -1):
        down_heapify(arr,i,n)
    #calling down_heapify from last index till 0
    for i in range(n-1,0,-1):
        #swapping 0 and last index element of arr as arr keeps smaller
        arr[0], arr[i] = arr[i], arr[0]
        down_heapify(arr,0,i)
    return

n = int(input())
arr = [int(ele) for ele in input().split()]
heapSort(arr)
for ele in arr:
    print(ele,end=' ')
```

```python
#K Smallest Elements

#importing heapq library
import heapq

def kSmallest(lst, k):
    #checking length of list
    if len(lst) == 0:
        return
    
    #getting first k element save in new list kSmallest
    kSmallest = lst[0:k]
    #max heapify on ksmallest list
    heapq._heapify_max(kSmallest)
    
    #traversing on lst from k to length of list
    for i in range(k,len(lst),1):
        #getting element at index i from lst
        ele = lst[i]
        #checking ele is small than max element in heap kSmallest
        if ele < kSmallest[0]:
            #replacing max element in kSmallest with ele
            heapq._heapreplace_max(kSmallest, ele)
            
    return kSmallest
    
# Main
n=int(input())
lst=list(int(i) for i in input().split())
k=int(input())
ans=kSmallest(lst, k)
ans.sort()
print(*ans, sep=' ')
```

```python
#K Largest Elements

#importing heapq library
import heapq
def kLargest(lst, k):
    #checking length of list
    if len(lst) == 0:
        return
    
    #getting first k element save in new list kSmallest
    kSmallest = lst[0:k]
    heapq.heapify(kSmallest)
    
    #traversing on lst from k to length of list
    for i in range(k,len(lst),1):
        #getting element at index i from lst
        ele = lst[i]
        #Checking ele is greater than min element in heap kSmallest
        if ele > kSmallest[0]:
            #replacing min element in kSmallest with ele
            heapq.heapreplace(kSmallest,ele)
    return kSmallest

# Main Code
n=int(input())
lst=list(int(i) for i in input().split())
k=int(input())
ans=kLargest(lst, k)
print(*ans, sep='\n')
```

```python
#Check Max-Heap

#importing heapq library
import heapq

def checkMaxHeap(lst):
    #checking length of list
    if len(lst) == 0:
        return
    #creating an empty list
    heapList = []
    #adding lst to heapList
    heapList.extend(lst)
    #Converting lst to max heap
    heapq._heapify_max(lst)
    #checking both list are identical or not
    if heapList == lst:
        return True
    else:
        return False

# Main Code
n=int(input())
lst=list(int(i) for i in input().split())

if checkMaxHeap(lst):
    print('true')
else:
    print('false')
```

```python
#Kth largest element

#importing heapq library
import heapq

def kthLargest(lst, k):
    #checking length of list
    if len(lst) == 0:
        return
    #Converting lst to max heap
    heapq._heapify_max(lst)
    #popping out max element from heap lst till i reaches to k
    for i in range(0,k,1):
        ele = heapq._heappop_max(lst)
    return ele

# Main
n=int(input())
lst=list(int(i) for i in input().split())
k=int(input())
ans=kthLargest(lst, k)
print(ans)
```

```python
#Buy the ticket

import sys
import heapq

class LinkedListNode :
    def __init__(self, data) :
        self.data = data
        self.next = None
        
class Queue :
    def __init__(self) :
        self.head = None 
        self.tail = None
        self.size = 0
        
    def enqueue(self, data) :
        newNode = LinkedListNode(data)
        if self.head is None :
            self.head = self.tail = newNode
        else :
            self.tail.next = newNode
            self.tail = newNode
        self.size += 1
        return
        
    def dequeue(self) :
        if self.head is None :
            return None
        data = self.head.data
        self.head = self.head.next
        self.size -= 1
        return data
    
    def getSize(self) :
        return self.size
    
    def isEmpty(self) :
        if self.head is None :
            return True
        return False
    
    def peek(self) :
        if self.head is None :
            return None
        return self.head.data
    
def buyTicket(arr, n, k):
    #if n is 0 then return
    if n == 0:
        return
    #making empty queue q
    q = Queue()
    #adding index of element in arr to q
    for i in range(0,len(arr),1):
        q.enqueue(i)
    #creating empty lis heapList
    heapList = []
    #adding values from arr to heapList
    heapList.extend(arr)
    #Converting heapList to max heap
    heapq._heapify_max(heapList)
    #making count as 0
    count = 0
    #till q is not empty
    while not q.isEmpty():
        #getting front element from queue
        front = q.dequeue()
        #checking if element at index front in arr is equal to max elemnt in heaplist
        if arr[front] == heapList[0]:
            #increasing count by 1
            count = count + 1
            #popping out max element from heap heapList
            heapq._heappop_max(heapList)
            #if front equal to k then breaking loop
            if front == k:
                break
        else:
            #adding back the front element into queue if it is not with max priority
            q.enqueue(front)
    return count        

#taking input using fast I/O
def takeInput() :
    n = int(input())
    if n == 0 :
        return n, list(), int()
    arr = list(int(i) for i in input().split())
    k = int(input())
    return n, arr, k

#main
sys.setrecursionlimit(10**6)
n, arr, k = takeInput()
print(buyTicket(arr, n, k))
```
