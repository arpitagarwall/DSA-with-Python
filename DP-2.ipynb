```python
#Min Cost Path Problem

from sys import setrecursionlimit
setrecursionlimit(10**6)
MAX_VALUE = 2147483647

def minCostPath(arr, i, j, mRows, nCols):
    #base case check if value of i and j is equal to 1 less than total rows and columns
    if i == mRows - 1 and j == nCols - 1:
        return arr[i][j]
    #special case if i and j become greater than or equal to total rows and columns then returning max value
    if i >= mRows or j >= nCols:
        return MAX_VALUE
    #Assuming induction hypothesis calling recurssion minCostPath with 3 different input value
    ans1 = minCostPath(arr, i+1, j, mRows, nCols)
    ans2 = minCostPath(arr, i, j+1, mRows, nCols)
    ans3 = minCostPath(arr, i+1, j+1, mRows, nCols)
    
    
    #calculatiing final answer by adding current value at index [i][j] in arr to min(ans1,ans2,ans3)
    ans = arr[i][j] + min(ans1,ans2,ans3)
    
    return ans
    
#taking input using fast I/O method
def take2DInput() :
    li = [int(ele) for ele in input().split()]
    mRows = int(li[0])
    nCols = int(li[1])
    
    if mRows == 0 :
        return list(), 0, 0
    
    mat = [[int(ele) for ele in input().split()]for row in range(mRows)]
    return mat, mRows, nCols


#main
mat, mRows, nCols = take2DInput()
print(minCostPath(mat, 0, 0, mRows, nCols))
```

```python
#LCS - Problem
def lcs(s, t, i, j,dp):
    #base case if i and j are equal to length of s and t then return 0
    if i == len(s) or j == len(t):
        return 0
    #if character at index i in s and character at index j in t are equal 
    if s[i] == t[j]:
        #we check in dp array at index[i+1][j+1] if answer not present in it
        if dp[i+1][j+1] == -1:
            #assuming induction hypothesis recursion will give give length consequence string
            out = lcs(s, t, i+1, j+1,dp)
            #adding that answer to index [i+1][j+1]
            dp[i+1][j+1] = out
        else:
            #if answer is present at dp[i+1][j+1] then stopping overlapping calls
            out = dp[i+1][j+1]
        #adding 1 to the answer because 1st call was made by user
        ans = 1 + out
        return ans
    #if character at index i in s and character at index j in t are not equal
    if s[i] != t[j]:
        #we check in dp array at index[i+1][j+1] if answer not present in it
        if dp[i+1][j] == -1:
            #assuming induction hypothesis recursion will give give length consequence string
            out1 = lcs(s, t, i+1, j,dp)
            #adding that answer to index [i+1][j]
            dp[i+1][j] = out1
        else:
            #if answer is present at dp[i+1][j] then stopping overlapping calls
            out1 = dp[i+1][j]
        #we check in dp array at index[i][j+1] if answer not present in it
        if dp[i][j+1] == -1:
            #assuming induction hypothesis recursion will give give length consequence string
            out2 = lcs(s, t, i, j+1,dp)
            #adding that answer to index [i][j+1]
            dp[i][j+1] = out2
        else:
            #if answer is present at dp[i][j+1] then stopping overlapping calls
            out2 = dp[i][j+1]
        #checking max  from both out1 and out2
        ans = max(out1,out2)
        return ans
#main
s = str(input())
t = str(input())
dp = [[-1 for i in range(len(t) + 1)] for j in range(len(s) + 1)]
print(lcs(s, t, 0, 0,dp))
```

```python
#0 1 Knapsack - Problem

def knapsack(weights, values, n, maxWeight,i):
    #base case if i == len(values) then return 0
    if i == len(values):
        return 0
    #if weight at index i is less than maxWeight
    if weights[i] <= maxWeight:
        #assuming induction hypothesis, recurssion will return max weight by not including value at ith index
        ans1 = knapsack(weights, values, n, maxWeight,i+1)
        #assuming induction hypothesis, recurssion will return max weight by including value at ith index
        ans2 = values[i] + knapsack(weights, values, n, maxWeight - weights[i],i+1)
        #finding max ans1 and ans2
        ans = max(ans1, ans2)
        return ans
    #if weight at index i is more than maxWeight
    if weights[i] > maxWeight:
        #assuming induction hypothesis, recurssion will return max weight by not including value at ith index
        ans = knapsack(weights, values, n, maxWeight,i+1)
        return ans


def takeInput() :
    n = int(input())

    if n == 0 :
        return list(), list(), n, 0

    weights = [int(i) for i in input().split()]
    values = [int(i) for i in input().split()]
    maxWeight = int(input())

    return weights, values, n, maxWeight


#main
weights, values, n, maxWeight = takeInput()

print(knapsack(weights, values, n, maxWeight,0))
```

```python
#Matrix Chain Multiplication

import sys

def mcm(p,i,j,dp):
    #base case if i == j means we dont have matrix to multiply
    if i == j:
        return 0
    #starting minValue with sys.maxsize
    minValue = sys.maxsize
    #going from i to j using for loop
    for k in range(i,j):
        #checking in dp array at index [i][k] if -1 is present or not
        if dp[i][k] == -1:
            #assuming induction hypothesis recurssion will give cost to multiply from i to k matrix
            ans1 = mcm(p,i,k,dp)
            #saving that answer in dp array at index [i][k]
            dp[i][k] = ans1
        else:
            #using the save answer from dp array
            ans1 = dp[i][k]
        #checking in dp array at index [i][k] if -1 is present or not    
        if dp[k+1][j] == -1:
            #assuming induction hypothesis recurssion will give cost to multiply from k+1 to j matrix
            ans2 = mcm(p,k+1,j,dp)
            #saving that answer in dp array at index [k+1][j]
            dp[k+1][j] = ans2
        else:
            ans2 = dp[k+1][j]
        #calculating cost at particular i, j and k
        mCost = p[i-1]*p[k]*p[j]
        #finding total cost by adding nas1, ans2 and mCost
        currValue = ans1 + ans2 + mCost
        #finding minimum of minValue and currValue
        minValue = min(minValue,currValue)
        
    return minValue
        
n=int(input())
p=[int(i) for i in input().split()]
dp = [[-1 for j in range(n+1)] for j in range(n+1)]
print(mcm(p,1,n,dp))
```
