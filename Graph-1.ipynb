```python
#Code : Get Path - BFS

import queue
from sys import setrecursionlimit
setrecursionlimit(10**6)

#Defining new class as Graph
class Graph:
    #init method to initialise the constructor and making adjMatrix for nVertices
    def __init__(self, nVertices):
        self.nVertices = nVertices
        self.adjMatrix = [[0 for i in range(nVertices)] for j in range(nVertices)]
        
    def addEdge(self, v1, v2):
        #changing 0 to 1 at index v1 and v2 to show edge is added
        self.adjMatrix[v1][v2] = 1
        self.adjMatrix[v2][v1] = 1
        
    def __getPathHelper(self, sv, ev, visited):
        #creating empty dictionary
        mapp = {}
        #creating new queue q
        q = queue.Queue()
        #cheking if sv and ev is smaller than v or not
        if sv <V and ev<V:
            #if at index[sv][ev] in adjMatrix is equal to 1 or not and sv == ev or not. 
            if self.adjMatrix[sv][ev] == 1 and sv == ev:
                ans = []
                #appending sv to ans
                ans.append(sv)
                return ans
            #putting sv to q
            q.put(sv)
            #marking sv true in visited array
            visited[sv] = True
        else:
            return
        #while q is not empty
        while q.empty() is False:
            #getting first element of q and assigning it to front
            front = q.get()
            #using for loop
            for i in range(self.nVertices):
                #if at index[front][i] in adjMatrix is equal to 1 and visited[i] is false. 
                if self.adjMatrix[front][i] == 1 and visited[i] is False:
                    #adding value in mapp as front and key as i
                    mapp[i] = front
                    #putting i in q
                    q.put(i)
                    #marking i true in visited array
                    visited[i] = True
                    #if i == ev
                    if i == ev:
                        ans = []
                        #appending ev to ans
                        ans.append(ev)
                        #assinging value of mapp[ev] in value
                        value = mapp[ev]
                        #while value is not equal to sv
                        while value != sv:
                            #appending value in ans
                            ans.append(value)
                            value = mapp[value]
                        #appending value in ans
                        ans.append(value)
                        return ans
    
        return []
        
        
    def getPath(self, sv, ev):
        #creating visited array to keep track of thr vertices that has been visited once
        visited = [False for i in range(self.nVertices)]
        return self.__getPathHelper(sv, ev, visited)
        
    def removeEdge(self, v1, v2):
        #checking if there are edges in adjMatrix or not
        if self.containEdge(v1,v2) is False:
            return
        #assigning at that index equal to 0
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0
        
    def containEdge(self, v1, v2):
        #returning true if adjMatrix at index v1 and v2 is greater than 0
        return True if self.adjMatrix[v1][v2] > 0 else False
    
    def __str__(self):
        #for printing adjMatrix
        return str(self.adjMatrix)

# Main
li = [int(ele) for ele in input().split()]
V = int(li[0])
E = int(li[1])

g = Graph(V)

for i in range(E) :
    arr = [int(ele) for ele in input().split()]
    fv = int(arr[0])
    sv = int(arr[1])
    g.addEdge(fv, sv)

Vi = [int(ele) for ele in input().split()]
v1 = int(Vi[0])
v2 = int(Vi[1])
ans = g.getPath(v1,v2)
if ans is not None:
    print(*ans, end=" ")
```
