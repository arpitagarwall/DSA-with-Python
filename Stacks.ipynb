```python
#Code : Stack Using LL
#Following is the structure of the node class for a Singly Linked List
class Node:

    def __init__(self, data):
        #calling node calss constructor on data and next
        self.data = data
        self.next = None


class Stack:

    #Define data members and __init__()

    def __init__(self):
        #calling stack class constructor on __head and __count
        self.__head = None
        self.__count = 0

    def push(self, data):
        #Implement the push(element) function
        #first creating node from data
        newNode = Node(data)
        #then if head is none then making newNode.next  = None or if head is not none the connecting newnod.next = head
        newNode.next = self.__head
        #then making head as newNode
        self.__head = newNode
        #incementing count by 1
        self.__count = self.__count + 1

        
    def pop(self):
        #Implement the pop() function
        #first checking ii is empty is true or not
        if self.isEmpty() is True:
            return -1
        #saving current head.data in data
        data = self.__head.data
        #changing current head to head.next
        self.__head = self.__head.next
        #decrementing count by 1
        self.__count = self.__count - 1
        return data
    
    def top(self):
        #Implement the top() function
        #first checking ii is empty is true or not
        if self.isEmpty() is True:
            return -1
         #saving current head.data in data which is top element
        data = self.__head.data
        return data

    def getSize(self):
        #Implement the getSize() function
        return self.__count


    def isEmpty(self):
        #Implement the isEmpty() function
        return self.getSize() == 0



#main
q = int(input())

stack = Stack()

while q > 0 :

    inputs = list(int(ele) for ele in input().split())
    choice = int(inputs[0])

    if choice == 1 :
        data = int(inputs[1])
        stack.push(data)

    elif choice == 2 :
        print(stack.pop())

    elif choice == 3 :
        print(stack.top())

    elif choice == 4 : 
        print(stack.getSize())

    else :
        if stack.isEmpty() :
            print("true")
        else :
            print("false")

    q = q - 1
```

```python
#Balanced Paranthesis
from sys import stdin

#isBlanced Function
def isBalanced(expression):
    #we can use list as a stack
    s = []
    #taking flag as false
    flag = False
    #iteraing on each element of string
    for i in expression:
        #if it is open bracket we will append it to list
        if i == "(":
            s.append(i)
        else:
            #if list is empty we return false
            if len(s) == 0:
                flag = False
            else:
                #else we pop last element of list if it is closed bracket
                s.pop()
                #setting flag true
                flag = True
    #at last if list is empty and flag is true then return true
    if len(s) == 0 and flag is True:
        return True
    else:
        #else return false
        return False


#main
expression = stdin.readline().strip()
if isBalanced(expression):
    print("true")
else:
    print("false")
```

```python
#Reverse Stack
from sys import setrecursionlimit
setrecursionlimit(10 ** 6)

#Function reverseStack()
def reverseStack(inputStack, extraStack):
    #first we calculate the length of stack
    sizeStack = len(inputStack)
    #then we handle the base case
    if sizeStack == 0:
        return
    #traversing through inputStack till sizeStack-1 index
    for i in range(0,sizeStack-1,1):
            #appending all value of inputStack to extraStack except last value
            extraStack.append(inputStack.pop())
    #saving last value in stack to element
    element = inputStack.pop()
    #traversing through extraStack
    for j in range(0,len(extraStack),1):
        #inserting all element of extraStack to inputStack
        inputStack.append(extraStack.pop())
    #Assuming induction hypothesis  that recurssion will give reverse stack of n- 1 size
    reverseStack(inputStack, extraStack)
    #Appending last element to the top of stack
    inputStack.append(element)
    
    return inputStack


#Takes a list as a stack and returns whether the stack is empty or not
def isEmpty(stack):
    return len(stack) == 0


#Taking input using fast I/o method
def takeInput():
    size = int(input())
    inputStack = list()

    if size == 0:
        return inputStack


    values = list(int(ele) for ele in input().split())

    inputStack = values

    return inputStack


# Main
inputStack = takeInput()
emptyStack = list()

reverseStack(inputStack, emptyStack)

while not isEmpty(inputStack):
    print(inputStack.pop(), end = " ")
```

```python
#Check redundant brackets
#function checkRedundantBrackets()
def checkRedundantBrackets(expression):
    #first we take an empty list and use it as astack
    s = []
    #for every character in expression
    for i in expression:
        #if i != ) then we append it to list
        if i != ")":
            s.append(i)
        else:
            #else we check th count how many characters are ther in list betwee brackets
            count = 0
            while s.pop() != "(":
                count = count + 1
            #if there are less then or eqau to one character the we return True else false    
            if count <= 1:
                return True
    return False
#main
expression = input()

if checkRedundantBrackets(expression):
    print("true")

else:
    print("false")
```

```python
#Stock Span
#function stockSpan()
def stockSpan(price, n):
    #we maintain 2 list stack and span
    #In stach we store index of the stock price
    #In Span we store Total Number of consecutive days stock prcie is higher than previous one
    stack = []
    span = []
    #base case we store first index = 0 in stack and first span = 1 in span
    stack.append(0)
    span.append(1)
    #trversing through every element in price
    for i in range(1,len(price),1):
        #In While we check current stock is greater than previous stock
        while price[i] > price[stack[-1]]:
            #we remove the index from the stack
            stack.pop()
            #if stack is empty then break the while loop
            if len(stack) == 0:
                break
        #if length of stack is greater than 0
        if len(stack) > 0:
            #then we append into stack (i-stack[-1])
            span.append(i - stack[-1])
        else:
            #else we append span with i + 1
            span.append(i + 1)
        #then we updat stack with current index of stock which is greater then previous ones
        stack.append(i)
    return span

def printList(arr):
    for i in range(len(arr)):
        print(arr[i], end = " ")
    print()


def takeInput():
    size = int(input())

    if size == 0 :
        return list(), 0

    price = list(int(ele) for ele in input().split())

    return price, size


#main
price, n = takeInput()

output = stockSpan(price, n)
printList(output)
```
