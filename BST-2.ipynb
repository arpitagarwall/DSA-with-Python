```python
#Find path in BST

import queue

class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def findPathBST(root,data):
    #base case if root is none then return none
    if root == None:
        return None
    #edge case if root.data == to data then create a list and append root.data into it
    if root.data == data:
        lst = []
        lst.append(root.data)
        return lst
    #we will compare the data with root.data if it is greater than root.data then we search in right suntree
    #else we will search in left subtree
    if root.data > data:
        #assuming induction hypothesis recurrsion will give list if data is present in left subtree or give None
        leftResult = findPathBST(root.left,data)
        #if leftResult is not None
        if leftResult != None:
            #we will append root.data in leftResult and return it
            leftResult.append(root.data)
            return leftResult
    elif root.data <= data:
        #assuming induction hypothesis recurrsion will give list if data is present in right subtree or give None
        rightResult = findPathBST(root.right,data)
        #if rightResult is not None
        if rightResult != None:
            #we will append root.data in rightResult and return it
            rightResult.append(root.data)
            return rightResult
    #if data is not present in tree we will return none
    else:
        return None
        

def buildLevelTree(levelorder):
    index = 0
    length = len(levelorder)
    if length<=0 or levelorder[0]==-1:
        return None
    root = BinaryTreeNode(levelorder[index])
    index += 1
    q = queue.Queue()
    q.put(root)
    while not q.empty():
        currentNode = q.get()
        leftChild = levelorder[index]
        index += 1
        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            currentNode.left =leftNode
            q.put(leftNode)
        rightChild = levelorder[index]
        index += 1
        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            currentNode.right =rightNode
            q.put(rightNode)
    return root

# Main
levelOrder = [int(i) for i in input().strip().split()]
root = buildLevelTree(levelOrder)
data = int(input())
path = findPathBST(root,data)
if path is not None:
    for ele in path:
        print(ele,end=' ')
```

```python
#BST Class
class BinaryTreeNode:
    
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
class BST:
    
    def __init__(self):
        self.root = None
        self.numNodes = 0
    
    def printTreeHelper(self, root):
        #base case
        if root == None:
            return
        #printing root.data
        print(root.data, end = ":")
        #first calling on left subtree and print left child of root
        if root.left != None:
            print("L:" + str(root.left.data), end = ",")
         #then calling on right subtree and print right child of root    
        if root.right != None:
            print("R:"+ str(root.right.data), end = "")
        #blank print to enter in next line
        print()
        #assuming induction hypothesis recurssion will go on every node of left and right subtree
        self.printTreeHelper(root.left)
        self.printTreeHelper(root.right)
        
    def printTree(self):
        #calling printTreeHelper because recurssion cannot be used in printTree
        self.printTreeHelper(self.root)
    
    def searchHelper(self,root,data):
        #base case
        if root == None:
            return False
        #if root.data == data return true
        if root.data == data:
            return True
        #camparing data with root.data if it is smaller then root.dat then call recurssion on left subtree
        if root.data > data:
            self.searchHelper(root.left,data)
        else:
            #else call recurssion on right subtree
            self.searchHelper(root.right,data)
        
    def search(self, data):
        #calling searchHelper because recurssion cannot be used in search
        return self.searchHelper(self.root,data)
    
    def insertHelper(self,root,data):
        #base case if root is none then we create new root using BinaryTreeNode class 
        if root == None:
            node = BinaryTreeNode(data)
            return node
        #if root.data is less than data than we call recurssion on right subtree and assign it to root.right
        if root.data < data:
            root.right = self.insertHelper(root.right,data)
            return root
        #else we call recurssion on left subtree and assign it to root.left
        else:
            root.left = self.insertHelper(root.left,data)
            return root
                  
    def insert(self, data):
        #calling insertHelper because recurssion cannot be used in insert
        self.root = self.insertHelper(self.root,data)
        #everytime when node is inserted we increase numNodes by 1
        self.numNodes = self.numNodes + 1
    
    def minRoot(self, root):
        #if root is none then return high value positive integer
        if root == None:
            return 100000
        #if root.left is none then return root
        if root.left == None:
            return root.data
        #assuming induction hypothesis recurssion will check all left child
        return self.minRoot(root.left)
    
    def deleteHelper(self, root, data):
        #base case if root is none then return false and None
        if root == None:
            return False, None
        #if data is smaller than root.data calling deleteHelper on left subtree and assigning newLeftroot to root.left
        if root.data > data:
            deleted, newLeftRoot = self.deleteHelper(root.left, data)
            root.left = newLeftRoot
            return deleted, root
        #if data is greater than root.data calling deleteHelper on right subtree and assigning newRightroot to root.right
        if root.data < data:
            deleted, newRightRoot = self.deleteHelper(root.right, data)
            root.right = newRightRoot
            return deleted, root
        #if root.left and root.right both are None means we have only root in tree then return True and None
        if root.left == None and root.right == None:
            return True, None
        #if root.left is none means there is no left child then return True and root.right
        if root.left == None:
            return True, root.right
        #if root.right is none means there is no right child then return True and root.left
        if root.right == None:
            return True, root.left
        #if root have both left and right child
        #assiginig minimum root from right subtree to replacement
        replacement = self.minRoot(root.right)
        #replacing root.data with replacement
        root.data = replacement
        #now we have to remove replacement from its original place
        deleted, newRightRoot = self.deleteHelper(root.right, replacement)
        #assigning newrightRoot to root.right
        root.right = newRightRoot
        
        return True, root
        
    def delete(self, data):
        #calling deleteHelper because recurssion cannot be used in delete
        deleted, newRoot = self.deleteHelper(self.root, data)
        #if deleted is true then decrease numNodes by 1
        if deleted:
            self.numNodes = self.numNodes - 1
        #assigning newRoot to self.root
        self.root = newRoot
        return deleted
    
    def count(self):
        return self.numNodes
        
b = BST()

q = int(input())

while (q > 0):
    
    li = [int(ele) for ele in input().strip().split()]
    
    choice = li[0]
    
    q-=1
    
    if choice == 1:
        
        data = li[1]
        b.insert(data)
        
    elif choice == 2:
        
        data = li[1]
        b.delete(data)
        
    elif choice == 3:
        
        data = li[1]
        ans = b.search(data)
        
        if ans is True:
            print('true')
        else:
            print('false')
            
    else:
        b.printTree()
```
