```python
#Eliminate duplicates from LL
#Following is the Node class already written for the Linked List
class Node:
    def __init__(self, data) :
        self.data = data
        self.next = None
        
def length(head):
    #we take variable count starting from 0
    count = 0
    #we will check the condition when head is not none we stop the loop 
    while head is not None:
        #we assign head with refrence of next number and incrementing count
        head = head.next
        count = count + 1
    return count
        
def removeDuplicates(head):
    #first we check the length of linked list 
    count = length(head)
    #if length is 0 then we return none
    if count == 0:
        return None
    #we take 2 variable prev and curr prev
    prev = head
    curr = head.next
    #traversing through linked list using while loop
    while curr is not None:
        #if prev.data and curr.data is equal then prev.next will refer to curr.next and curr will refer to curr.next
        if prev.data == curr.data:
            prev.next = curr.next
            curr = curr.next
        else:
            #else we assign curr to prev and curr.next to curr
            prev = curr
            curr = curr.next
    return head

#Taking Input Using Fast I/O
def takeInput():
    head = None
    tail = None

    datas = list(int(ele) for ele in input().split())

    i = 0
    while (i < len(datas)) and (datas[i] != -1):
        data = datas[i]
        newNode = Node(data)

        if head is None:
            head = newNode
            tail = newNode

        else:
            tail.next = newNode
            tail = newNode

        i = i + 1

    return head

#to print the linked list 
def printLinkedList(head):

    while head is not None:
        print(head.data, end = " ")
        head = head.next
    print()

#main
t = int(input())

while t > 0:
    head = takeInput()
    head = removeDuplicates(head)
    printLinkedList(head)
    t = t - 1
```
