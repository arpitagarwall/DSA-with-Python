```python
#Maximum Frequency
def maxfreq(arr):
    #Maintaing a empty dictionary d
    d={}
    #adding elements from arr to dictionary d
    for i in arr:
        d[i] = d.get(i,0) + 1
        
    #maintainig variable max and assigning value of arr[0] to it
    max=arr[0]
    #looping through arr checking if d[i] > d[max] then updating max to i
    for i in arr:
        if d[i] > d[max]:
            max=i
    return max
    
def takeInput():
    #To take fast I/O
    n = int(input())
    
    if n == 0:
        return list(), 0
    
    arr = [int(ele) for ele in input().split()]
    return arr,n

arr, n = takeInput()
print(maxfreq(arr))
```

```python
#Pair Sum To 0
from sys import setrecursionlimit
from math import factorial
setrecursionlimit(10 ** 6)

#nc2 will be used to decide how many times pair(0,0) is to be printed
def nc2(n):
    #maintaining variable ans = 0
    ans = 0
    #if n not equal to 1
    if n != 1:
        #using the formula of combination
        ans = factorial(n) / (factorial(2)*factorial(n-2))
    #returning int as ans
    return int(ans)

#For creating frequency map
def freqMap(l):
    #maintaining variable map as dictionary
    map = {}
    #Using for loop to trverse on list
    for num in l:
        #if digit exist before in dictionary then increse its count to 1
        if num in map:
            map[num]+=1
        else:
            #else set it to 1
            map[num]=1
    return map


def pairSum0(l,n):
    #makiing dictionary of array
    m = freqMap(l)
    #setting count = 0
    count = 0
    #checking all keys in dictionary
    for num in m:
        #if key is greater than 0 and negative of key is present in dictionary
        if num > 0 and -num in m:
            #multiplying both the values of key and adding to count
            count = count + (m[num]*m[-num])
        #if num = 0 the calling nc2 to find combination and adding to count
        if num==0:
            count = count + nc2(m[num])
    return count

def takeInput():
    #To take fast I/O
    n = int(input())
    
    if n == 0:
        return list(), 0
    
    arr = [int(ele) for ele in input().split()]
    return arr,n

arr,n=takeInput()
print(pairSum0(arr,n))
```

```python
#Extract Unique characters
def uniqueChar(s):
    #maintaining variable ans for storing output string
    ans = '' 
    #creating empty dictionary d
    d = {} 
    #traversing on s
    for char in s:
        #checking if char is present in dictionary d or not
        if char not in d:
            #concatenating ans with char
            ans = ans + char
            #adding value true in front of key-char in dictionary d
            d[char] = True
    return ans 

# Main 
s=input() 
print(uniqueChar(s))
```

```python
#Longest Consecutive Sequence
def longestConsecutiveSubsequence(arr,n):
    #maintaining visitedElements and elementToIndexMapping as dictionary
    visitedElements = {}
    elementToIndexMapping = {}
    #using for loop in range(n)
    for i in range(n):
        #adding in dictionary elementToIndexMapping with key as arr[i] and value as i
        elementToIndexMapping[arr[i]] = i
        #if arr[i] not in dictionary visitedElements then adding into it with key as arr[i] and value as true
        if arr[i] not in visitedElements:
            visitedElements[arr[i]] = True
    #maintaining variable longestSequence as list, maxSequenceLength as int, minStartIndex as int       
    longestSequence=[]
    maxSequenceLength = 1
    minStartIndex = 0
    #using for loop in range(n)
    for i in range(n):
        num = arr[i]
        currentMinStartIndex = i
        count = 0
        tempNum = num
        #checking tempNum in visitedElements and checking tempNum value as True
        while tempNum in visitedElements and visitedElements[tempNum] == True:
            #changing its value to false
            visitedElements[tempNum] = False
            #increasing count by 1
            count += 1
            #increasing tempNum by 1 ti check consecutive element
            tempNum += 1
        #assigning tempNum to num - 1    
        tempNum = num-1
        #checking tempNum in visitedElements and checking tempNum value as True
        while tempNum in visitedElements and visitedElements[tempNum]==True:
            #changing its value to false
            visitedElements[tempNum ] = False
            #increasing count by 1
            count+=1
            #setting currentMinStartIndex to elementToIndexMapping[tempNum]
            currentMinStartIndex = elementToIndexMapping[tempNum]
            #decreasing tempnum by 1
            tempNum -=1
        #if count is greater than maxSequenceLength
        if (count > maxSequenceLength):
            #assigning maxSequenceLength to count
            maxSequenceLength = count
            #assigning minStartIndex to currentMinStartIndex
            minStartIndex = currentMinStartIndex
        #if count is equal to maxSequenceLength
        elif (count == maxSequenceLength):
            #if currentMinStartIndex is less than minStartIndex
            if(currentMinStartIndex < minStartIndex):
                minStartIndex = currentMinStartIndex
    #assigning start key to startNum
    startNum = arr[minStartIndex]
    #appending startNum to longestSequence
    longestSequence.append(startNum)
    #maxSequenceLength is greater than 1
    if (maxSequenceLength > 1):
        #appending (startNum + maxSequenceLength - 1) to longestSequence
        longestSequence.append(startNum + maxSequenceLength-1)     
    return longestSequence
                
def takeInput():
    #To take fast I/O
    n=int(input())
    if n==0:
        return list(),0
    arr=list(int(ele) for ele in input().split())
    return arr,n

    
# Main 
arr,n=takeInput()
ans = longestConsecutiveSubsequence(arr,n) 
# This ans array contains two numbers, ie, start and end of longest sequence respectively
print(*ans)
```

```python
#Pairs with difference K
def printPairDiffK(l, k):
    #maintainind dictionary d
    d = {}
    #traversing on list l and and adding it to dictionary d
    for ele in l:
        d[ele] = d.get(ele, 0)+1
    #maintainign count of pair
    count = 0
    #traversing on dictionary
    for ele in d:
        #checking if k + ele in d than increasing count by multiplying the keys values
        if k+ele in d:
            count = count + d[ele]*d[ele+k]
        #if dictionary has same key and same value with length 1 then using formula given below      
        if len(d) == 1:
            count = d[ele]*(d[ele]-1)//2
            return count
    return count
    
# Main
n=int(input())
l=list(int(i) for i in input().split())
k=int(input())
print(printPairDiffK(l, k))
```

```python
#Longest subset zero sum
def subsetSum(l):
    #base case if length of l is 0 then return 0
    if len(l) == 0:
        return 0
    #maintainig variables maxSum, max_length equal to 0 and a_subset as dictionary
    maxSum = 0 
    max_length = 0
    a_subset = {}
    #traversing on list l
    for i in range(len(l)):
        #increasing maxSum by adding list element at i
        maxSum = maxSum + l[i]
        #if maxSum is equal to 0 then adding 1 in index and saving it
        if maxSum == 0:
            val = i + 1
            #if val is greater than max_length
            if val > max_length:
                #assigning val to max_length
                max_length = val
        #checking if maxsum in a_subset
        elif maxSum in a_subset:
            #decreasing val by subtracting value of key from i
            val = i - a_subset[maxSum]
            if val > max_length:
                #checking max_length and assigning to it
                max_length = val
            continue
        #adding maxSum to dictionary with value i
        a_subset[maxSum] = i    
            
    return max_length    

# Main
n=int(input())
l=list(int(i) for i in input().split())
print(subsetSum(l))
```
