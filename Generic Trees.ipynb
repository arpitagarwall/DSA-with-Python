```python
#Sum of all nodes
import sys
import queue

class TreeNode :
    def __init__(self, data) :
        self.data = data
        self.children = list()

def inputLevelWise(li) :
    i = 0
    data = li[i] 
    i += 1
    
    if data == -1 :
        return None
    
    root = TreeNode(data) 
    q = queue.Queue()
    q.put(root)
    
    while (not q.empty()):
        
        frontNode = q.get()
        noOfChildren = li[i]
        i += 1
        
        childrenArray = li[i : i+noOfChildren]
        
        for childData in childrenArray :
            childNode = TreeNode(childData)
            frontNode.children.append(childNode)
            q.put(childNode)
            
        i = i+noOfChildren
        
    return root
        
def sumOfAllNodes(root):
    #edge case if root i none then return 0
    if root == None:
        return 0
    #maintaning sumNodes and assigning forst value as root.data
    sumNodes = root.data
    #using for loop to go to every children in root
    for child in root.children:
        #assuming induction hypothesis recurssion will sum of all nodes except root
        sumNodes = sumNodes + sumOfAllNodes(child)
    #returning sumNodes
    return sumNodes
    
#main
sys.setrecursionlimit(10**6)
li = [int(elem) for elem in list(input().strip().split())]
root = inputLevelWise(li)
print(sumOfAllNodes(root))
```

```python
#Height Of Tree
from sys import stdin
import sys
import queue

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def heightOfTree(root):
    #if root is none return 0
    if root == None:
        return 0
    #maintaing variable height as 0
    height = 0
    #using for loopto go through each children of every root
    for child in root.children:
        #assuming induction hypothesis recurssion will give height of tree except root
        finalHeight = heightOfTree(child)
        #finding which child hase maximum height
        height = max(height, finalHeight)
    #returning height and adding 1 for root in it
    return height + 1
        
def inputLevelWise(li) :
    i = 0
    data = li[i] 
    i += 1
    
    if data == -1 :
        return None
    
    root = TreeNode(data) 
    q = queue.Queue()
    q.put(root)
    
    while (not q.empty()):
        
        frontNode = q.get()
        noOfChildren = li[i]
        i += 1
        
        childrenArray = li[i : i+noOfChildren]
        
        for childData in childrenArray :
            childNode = TreeNode(childData)
            frontNode.children.append(childNode)
            q.put(childNode)
            
        i = i+noOfChildren
        
    return root

# Main
sys.setrecursionlimit(10**6)
li = [int(elem) for elem in list(input().strip().split())]
root = inputLevelWise(li)
print(heightOfTree(root))
```

```python
#Node with maximum child sum
from sys import setrecursionlimit
import sys
import queue
setrecursionlimit(10**6)

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    
def maxSumNode(root):
    #if root is none then return root as none and sum as 0
    if root == None:
        return None, 0
    #maintaining outputNode as root
    outputNode = root
    #maintaining sumChild as 0
    sumChild = 0
    #using for loop to sum of all child.data present in root
    for child in root.children:
        sumChild = sumChild + child.data
    
    #using for loop to go through every node in tree
    for child in root.children:
        #assuming induction hypothesis recurssion will give node and sum of all its child
        node, nodeSum = maxSumNode(child)
        #finding max sum of child of particular node
        sumChild = max(sumChild, nodeSum)
        #if sumChild == nodeSum than update outputNode
        if sumChild == nodeSum:
            outputNode = node
    #returning outputNode and sumChild
    return outputNode,sumChild
    
    

def inputLevelWise(li) :
    i = 0
    data = li[i] 
    i += 1
    
    if data == -1 :
        return None
    
    root = TreeNode(data) 
    q = queue.Queue()
    q.put(root)
    
    while (not q.empty()):
        
        frontNode = q.get()
        noOfChildren = li[i]
        i += 1
        
        childrenArray = li[i : i+noOfChildren]
        
        for childData in childrenArray :
            childNode = TreeNode(childData)
            frontNode.children.append(childNode)
            q.put(childNode)
            
        i = i+noOfChildren
        
    return root

# Main
li = [int(elem) for elem in list(input().strip().split())]
root = inputLevelWise(li)
node, sumChild = maxSumNode(root)
print(node.data)
```

```python
#Structurally identical
from sys import setrecursionlimit
import sys
import queue
setrecursionlimit(10**6)

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    
def isIdentical(root1, root2):
    #maintaining variable final as false
    final = False
    
    #checking edge case if root1 or root2 is none then return
    if root1 == None or root2 == None:
        return final
    
    #comapring data of both root and if they are equal then assign final as true else false
    if root1.data == root2.data:
        final = True
    else:
        final = False
    
    #maintainig 2 variable lenght1 and lenght2 
    lenght1 = len(root1.children)
    lenght2 = len(root2.children)
    
    #if lenght1 and lenght2 are equal then return true else false
    if lenght1 == lenght2:
        final = True
    else:
        final = False
    
    #using for loop simultaneuosly on 2 root children list using zip
    for (child1, child2) in zip(root1.children, root2.children):
        #assuming induction hypothesis recursion will return true or false
        output = isIdentical(child1, child2)
        #if output is true then final = true else final = false
        if output:
            final = True
        else:
            final = False
            
    return final
    
def inputLevelWise(li) :
    i = 0
    data = li[i] 
    i += 1
    
    if data == -1 :
        return None
    
    root = TreeNode(data) 
    q = queue.Queue()
    q.put(root)
    
    length = len(li)
    
    if length > 1:
        while (not q.empty()):

            frontNode = q.get()
            noOfChildren = li[i]
            i += 1

            childrenArray = li[i : i+noOfChildren]

            for childData in childrenArray :
                childNode = TreeNode(childData)
                frontNode.children.append(childNode)
                q.put(childNode)

            i = i+noOfChildren
        
    return root

# Main
li1 = [int(elem) for elem in list(input().strip().split())]
root1 = inputLevelWise(li1)
li2 = [int(elem) for elem in list(input().strip().split())]
root2 = inputLevelWise(li2)
if isIdentical(root1, root2):
    print('true')
else:
    print('false')
```

```python
#Next larger
from sys import setrecursionlimit
import sys
import queue
setrecursionlimit(10**6)

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    
def nextLargest(root, n):
    #edge case if root is none then return none
    if root == None:
        return None
    #maintaing variable greaterThanN setting it to None initially
    greaterThanN = None
    #then we check if root.dat is greater than n the we assign root to greaterThanN
    if root.data > n:
        greaterThanN = root
    #using for loop to go through each child of root
    for child in root.children:
        #assumg induction hypothesis recurssion will give smallest node from subtree
        output = nextLargest(child, n)
        #if greaterThanN is None then assign output to it becuase we found greater than n in subtrees
        if greaterThanN == None:
            greaterThanN = output
        #if output is not none and output is smaller than greaterThanN than we update greaterThanN to output
        #so that we get smallest node
        if output != None:    
            if output.data < greaterThanN.data:
                greaterThanN = output
        
    return greaterThanN
            
def inputLevelWise(li) :
    i = 0
    data = li[i] 
    i += 1
    
    if data == -1 :
        return None
    
    root = TreeNode(data) 
    q = queue.Queue()
    q.put(root)
    length = len(li)
    
    if length > 1:
        while (not q.empty()):

            frontNode = q.get()
            noOfChildren = li[i]
            i += 1

            childrenArray = li[i : i+noOfChildren]

            for childData in childrenArray :
                childNode = TreeNode(childData)
                frontNode.children.append(childNode)
                q.put(childNode)

            i = i+noOfChildren
        
    return root

# Main
li = [int(elem) for elem in list(input().strip().split())]
root = inputLevelWise(li)
n = int(input())
node = nextLargest(root, n)
if node != None:
    print(node.data)
```

```python
#Replace with depth
from sys import setrecursionlimit
import sys
import queue
setrecursionlimit(10**6)

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def replacewithDepth(root, level = 0):
    #if root is none then return
    if root == None:
        return
    #replacing value of root.data with level
    root.data = level
    #using for loop to iterate on each children
    for child in root.children:
        #assuming induction hypothesis recurssion will replace data of all the nodes in subtree with it's level
        replacewithDepth(child, level + 1)
    
def inputLevelWise(li) :
    i = 0
    data = li[i] 
    i += 1
    
    if data == -1 :
        return None
    
    root = TreeNode(data) 
    q = queue.Queue()
    q.put(root)
    length = len(li)
    
    if length > 1:
        while (not q.empty()):

            frontNode = q.get()
            noOfChildren = li[i]
            i += 1

            childrenArray = li[i : i+noOfChildren]

            for childData in childrenArray :
                childNode = TreeNode(childData)
                frontNode.children.append(childNode)
                q.put(childNode)

            i = i+noOfChildren
        
    return root

def printLevelAtNewLine(root):
    q = [root]
    newq = []
    while q:
        parent = q.pop(0)
        print(parent.data, end=' ')
        for child in parent.children:
            newq.append(child)
        if len(q)==0:
            q = newq
            newq = []
            print()  # Move to next Line

# Main
li = [int(elem) for elem in list(input().strip().split())]
root = inputLevelWise(li)
replacewithDepth(root)
printLevelAtNewLine(root)
```
